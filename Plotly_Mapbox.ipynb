```python
def plotly_temp_plot(data,date_,save_path,column_name):
    df = data.loc[data['day'] == date_]
    fig = px.scatter_mapbox(df, lat="lat", lon="lon",color = column_name,color_continuous_scale=px.colors.cyclical.IceFire, zoom=2.5, height=300)
    fig.update_layout(mapbox_style="open-street-map")
    fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
    fig.write_image(save_path)
```
